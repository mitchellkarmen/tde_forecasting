```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
stellar_mass = np.linspace(10**(8.5), 10**(11.5), 50)
x = stellar_mass/1e10

A = 0.91 + 0.77*x + 0.11*(x**2) - 0.09*(x**3)
```

```python
plt.plot(x, A)
plt.plot(x, -0.09*(x**3))
plt.show()
```

```python
def A_lambda_to_A_V(A_lambda, lbda, R_V):
    x = 1/lbda
    if 0.63 <= lbda <= 2.20: #micro meters
        coeff = ((2.659*(-1.857 + 1.040*x))/R_V) + 1
    elif 0.12 < lbda < 0.63:
        coeff = (2.659*(-2.156 + (1.509/lbda) - (0.198/lbda**2) + (0.011/lbda**3))/R_V) + 1
    return A_lambda/coeff
```

```python
h_alpha = 0.656281 #micro m
```

```python
masses = np.linspace(1e9, 10**(10.7))
A_Vs = []

for m_stellar in masses:
    x = m_stellar/1e10

    A_lambda = 0.91 + 0.77*x + 0.11*(x**2) - 0.09*(x**3)
    
    A_V = A_lambda_to_A_V(A_lambda, h_alpha, R_V=4.2)
    A_Vs.append(A_V)
```

```python
plt.plot(np.log10(masses), A_Vs)
plt.xlabel('log$_{10}$(host M$_*$/M$_\\odot$)')
plt.ylabel("A$_V$")
```

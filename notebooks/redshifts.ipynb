```python
TDE_names = list(gezari.Name.values)
combined = '\t'.join(TDE_names)

for hammer_name in TDEs.IAUName.values:
    if hammer_name in combined:
        print(hammer_name, ' already in table')
    else:
        TDE_names.append(hammer_name)
```

```python
len(TDE_names)
```

```python
gezari.columns
```

```python
redshifts = []

for name in TDE_names:
    if name in gezari.Name.values:
        z = gezari[gezari.Name == name][' Redshift']
        redshifts.append(float(z))
    elif name in TDEs.IAUName.values:
        z = TDEs[TDEs.IAUName == name]['Redshift']
        redshifts.append(float(z))
    else:
        print('could not find', name)
```

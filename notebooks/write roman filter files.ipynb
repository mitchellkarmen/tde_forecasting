```python
import numpy as np
import pandas as pd
```

```python
filters = pd.read_csv('../data/Roman_effarea_20201130.csv', skiprows=1, delimiter = ',')
filters
```

```python
string_head = """# Units:
# \t "lambda" is in Angstroms
# \t "pass" is is the contribution to the detector signal per photon \n
"""
# insert bandpass name here
string_body = """
# \t Instrument: Roman Wide Field Instrument \n
"""

typedef_string = """
typedef struct {
  double lambda;
  double pass;
} KFILTER; \n
"""
```

```python
print(string_head+ string_body)
```

```python
roman_filters = [' F062', ' F087', ' F106', ' F129', ' F184', ' F146', ' F213']
for col in roman_filters:
    print(col)
    filter_model = filters[['Wave', col]]
    #filter_model = filter_model.rename(columns={col: col[1:]})
    filter_string = '# \t Bandpass Name: {}'.format(col[1:])
    
    doc_string = string_head + filter_string + string_body + typedef_string
    
    total_string = doc_string
    for i, row in filter_model.iterrows():
        wave = row['Wave']*10000
        response = row[col]/ wave
        print(wave, row[col])
        row_string = 'KFILTER {:0.0f}. {}\n'.format(wave, response)
        total_string += row_string
        
    #print(total_string)
    file_path = '/Users/mitchell/opt/anaconda3/lib/python3.9/site-packages/kcorrect/data/responses/roman_{}_dwave.par'.format(col[1:])
    with open(file_path, "w") as f:
        f.write(total_string)
    
    #filter_model.to_csv('../data/roman_filters/roman_{}.dat'.format(col[1:]), sep=' ', index=False)
```
